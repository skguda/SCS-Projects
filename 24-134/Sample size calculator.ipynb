```python
import numpy as np
import pandas as pd
from scipy.stats import norm
```

## - Power = 0.8 
- Effect size = [0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1]
- Alpha = [0.01, 0.05]

$n = (\frac{2*(Z_{\alpha/2} +Z_{\beta})^2}{(Effect Size)^2})$

Assumption: Variance = 1

```python
es = [0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1]
power = 0.8
alpha = [0.01,0.05]


n1 = []
n5 = []
for i in es:
    v1 = 2*((norm.ppf(1-alpha[0]/2) + norm.ppf(power))/i)**2
    v5 = 2*((norm.ppf(1-alpha[1]/2) + norm.ppf(power))/i)**2
    
    n1.append(v1)
    n5.append(v5)

effectDF = pd.DataFrame({"Effect Size":es,"N|alpha=0.01":n1, "N|alpha = 0.05":n5})
effectDF
```
